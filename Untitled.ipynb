```python
import requests
import numpy as np
import pandas as pd
```

```python
start = 'https://lichess.org/api/games/user/'
finish = '?rated=true&analysed=false&tags=true&clocks=false&evals=false&opening=true&since=1579813200000&until=1658696400000&perfType=blitz%2Crapid'
def downloadDB (nickname):
    req = requests.get(start+nickname+finish)
    base = str(req.content).split('\\n\\n\\n')
    base.pop(-1)
    ind=0
    for elem in base:
        temp = elem.translate({ord("["):None,ord("]"):None}).split('\\n')
        temp[0] = temp[0].replace("b\'","")
        for i in range(17):
            if " " in temp[i]:
                temp[i] = temp[i][temp[i].index(" ")+2:-1]
        temp[15] = temp[15].translate({ord("\\"):None })
        temp[5] = temp[5].replace("1/2","0.5")
        temp[5] = temp[5][:temp[5].index("-")]
        base[ind] = temp
        ind+=1
        
    return pd.DataFrame(base, columns=['Event','Site','Date','White','Black','Result','UTCDate','UTCTime','WhiteElo','BlackElo', 'WhiteRatingDiff',
                                'BlackRatingDiff','Variant','TimeControl','ECO','Opening','Termination','What?','Moves'])
    
```

```python
df = downloadDB("ViKarp")
df.head
```

```python
openings = set(df["Opening"])
openings
```

```python
df["MatchesWhite"]=[0]*df.shape[0]
df["MatchesBlack"]=[0]*df.shape[0]
df["MatchesOpenWhite"] = [0]*df.shape[0]
df["MatchesOpenBlack"] = [0]*df.shape[0]
df["WinMatchesOpenWhite"] = [0]*df.shape[0]
df["WinMatchesOpenBlack"] = [0]*df.shape[0]
dict_openings = dict()
for j in range(df.shape[0]):
    df1 = downloadDB(df["White"][j])
    openings = set(df1["Opening"])
    df["MatchesWhite"]=df1.shape[0]
    for i in openings:
        dict_openings[i]=df1[df1["Opening"]==i].shape[0]
    for i in range(df1.shape[0]):
        df["MatchesOpenWhite"][i] = dict_openings[df1["Opening"][i]]
    win = 0
    matches = 0
    df2 = df1[df1["Opening"] == df["Opening"][j]]
    for i in df2.index:
        if (df2["Result"][i] == "1" or df2["Result"][i] == "0.5") and df2["White"][i] == df["White"][j]:
            win += float(df2["Result"][i])
        if df2["White"][i] == df["White"][j]:
                matches+=1
    df["WinMatchesOpenWhite"][i] = round(win/matches,2)    
    df1 = downloadDB(df["Black"][j])
    print(df1)
    openings = set(df1["Opening"])
    df["MatchesBlack"]=df1.shape[0]
    for i in openings:
        dict_openings[i]=df1[df1["Opening"]==i].shape[0]
    for i in range(df1.shape[0]):
        df["MatchesOpenBlack"][i] = dict_openings[df1["Opening"][i]]
    win = 0
    matches = 0
    df2 = df1[df1["Opening"] == df["Opening"][j]]
    
    
    for i in df2.index:
        if (df2["Result"][i] == "1" or df2["Result"][i] == "0.5") and df2["Black"][i] == df["Black"][j]:
            win += float(df2["Result"][i])
        if df2["Black"][i] == df["Black"][j]:
            matches+=1
    
    
    df["WinMatchesOpenBlack"][i] = round(win/matches,2)
```
